```julia
using CSV, DataFrames, Plots
```

```julia
df = CSV.File("../Data/pokemon.csv") |> DataFrame;
```

```julia
head(df,3)
```

```julia
mini_df = df[[:type1,:type2,:hp,:attack,:defense,:name]]
head(mini_df,3)
```

```julia
function type_df(atype)
    filter(row->row[:type1]==atype || row[:type1]==atype,mini_df)
end;
```

```julia
head(type_df("normal"))
```
